```python
import pydicom as dicom
import PIL # optional
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

# specify your image path
image_path = '/content/Patient_12.dcm'
ds = dicom.dcmread(image_path)
yo = ds.pixel_array
plt.imshow(yo)
im = Image.fromarray(yo)
#im.mode = 'I'
im.point(lambda i:i*(1./4)).convert('L').save('my.jpeg')
plt.show()
```
